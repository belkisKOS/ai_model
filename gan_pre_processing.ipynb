```python
import os
import shutil

# Define paths for the two classes
dataset_path = "/content/drive/My Drive/original_dataset/original"

pneumonia_folder = os.path.join(dataset_path, "Pneumonia")
pneumonia_viral_folder = os.path.join(dataset_path, "Pneumonia-Viral")

# Ensure that the Pneumonia folder exists
os.makedirs(pneumonia_folder, exist_ok=True)

# Move all images from "Pneumonia-Viral" to "Pneumonia"
for filename in os.listdir(pneumonia_viral_folder):
    src = os.path.join(pneumonia_viral_folder, filename)
    dst = os.path.join(pneumonia_folder, filename)
    shutil.move(src, dst)  # Move file from Pneumonia-Viral to Pneumonia

print(f"Successfully combined 'Pneumonia-Viral' into 'Pneumonia'.")
```

```python

from PIL import Image

# Define dataset path
dataset_path = "/content/drive/My Drive/original_dataset/original"
convert_categories = ["Pneumonia", "Lung_Opacity"]  # Only convert these categories

for category in convert_categories:
    folder_path = os.path.join(dataset_path, category)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if filename.endswith(".jpeg") or filename.endswith(".jpg"):  # Check for JPEG images
            img = Image.open(file_path).convert("RGB")  # Open image and convert to RGB
            new_file_path = file_path.rsplit(".", 1)[0] + ".png"  # Change extension to .png
            img.save(new_file_path, "PNG")  # Save as PNG

            # Check if the PNG file was created successfully
            if os.path.exists(new_file_path):
                print(f"Successfully converted {filename} to PNG")
            else:
                print(f"Failed to convert {filename} to PNG")

            os.remove(file_path)  # Remove the old JPEG file
            print(f"Removed original JPEG image: {filename}")
```

```python
import cv2

# Define dataset path
dataset_path = "/content/drive/My Drive/original_dataset/original"
target_size = (224, 224)  # Standard image size

for category in ["Normal", "Lung_Opacity", "Pneumonia"]:  # Updated class names
    folder_path = os.path.join(dataset_path, category)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        img = cv2.imread(file_path)

        if img is not None:
            img_resized = cv2.resize(img, target_size)  # Resize the image to 224x224
            cv2.imwrite(file_path, img_resized)  # Save the resized image back to the same path
            print(f"Resized {filename} to {target_size}")
        else:
            print(f"Failed to read {filename}")
```

```python
import numpy as np
import cv2

# Define dataset path
dataset_path ="/content/drive/My Drive/original_dataset/original"

for category in ["Normal", "Lung_Opacity", "Pneumonia"]:
    folder_path = os.path.join(dataset_path, category)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        img = cv2.imread(file_path)

        if img is not None:
            img_normalized = img / 255.0  # Normalize to range 0-1
            cv2.imwrite(file_path, (img_normalized * 255).astype(np.uint8))  # Save back as uint8
            print(f"Normalized {filename}")
        else:
            print(f"Failed to read {filename}")
```
