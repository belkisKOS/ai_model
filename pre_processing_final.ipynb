```python
#change to png
from PIL import Image


dataset_path = "/content/drive/My Drive/Atelier2025_SCI_1st_Dataset/SCI_DL_Workshop_XrayChest_Dataset"
convert_categories = ["Pneumonia", "Lung_Opacity"]

for category in convert_categories:
    folder_path = os.path.join(dataset_path, category)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if filename.endswith(".jpeg") or filename.endswith(".jpg"):
            img = Image.open(file_path).convert("RGB")
            new_file_path = file_path.rsplit(".", 1)[0] + ".png"
            img.save(new_file_path, "PNG")


            if os.path.exists(new_file_path):
                print(f"Successfully converted {filename} to PNG")
            else:
                print(f"Failed to convert {filename} to PNG")

            os.remove(file_path)  # Remove the old JPEG file
            print(f"Removed original JPEG image: {filename}")
```

```python
import random
import cv2
import os
import numpy as np

# Define dataset path
dataset_path = "/content/drive/My Drive/Atelier2025_SCI_1st_Dataset/SCI_DL_Workshop_XrayChest_Dataset"
augment_classes = ["Lung_Opacity"]

for category in augment_classes:
    folder_path = os.path.join(dataset_path, category)
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        img = cv2.imread(file_path)

        if img is not None:

            if random.random() > 0.5:
                img = cv2.flip(img, 1)
            if random.random() > 0.5:
                angle = random.randint(-15, 15)
                h, w = img.shape[:2]
                M = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1)
                img = cv2.warpAffine(img, M, (w, h))
            if random.random() > 0.5:
                brightness = random.uniform(0.8, 1.2)
                img = np.clip(img * brightness, 0, 255).astype(np.uint8)

            new_filename = f"aug_{filename}"
            cv2.imwrite(os.path.join(folder_path, new_filename), img)
            print(f"Augmented {new_filename}")
```

```python
#SPLIT THE DATA SET 70t/20v/10t

import os
import shutil
from sklearn.model_selection import train_test_split


dataset_path = "/content/drive/My Drive/Atelier2025_SCI_1st_Dataset/SCI_DL_Workshop_XrayChest_Dataset"
output_dir = "/content/drive/My Drive/Atelier2025_SCI_1st_Dataset/processed_dataset2"


for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(output_dir, split), exist_ok=True)


categories = ["Normal", "Lung_Opacity", "Pneumonia"]


for category in categories:
    category_path = os.path.join(dataset_path, category)


    all_images = os.listdir(category_path)


    train_files, test_files = train_test_split(all_images, test_size=0.3, random_state=42)


    val_files, test_files = train_test_split(test_files, test_size=0.33, random_state=42)


    for split, files in zip(["train", "val", "test"], [train_files, val_files, test_files]):
        split_folder = os.path.join(output_dir, split, category)
        os.makedirs(split_folder, exist_ok=True)


        for file in files:
            src = os.path.join(category_path, file)
            dst = os.path.join(split_folder, file)
            shutil.copy(src, dst)

        print(f"Moved {len(files)} images to {split}/{category}")

print("Dataset has been split successfully into train, val, and test sets.")
```
